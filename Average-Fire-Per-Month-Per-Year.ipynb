```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)

# Connect to the SQLite database file
conn = sqlite3.connect("FPA_FOD_20170508.sqlite")

# Read a table into a Pandas DataFrame (replace `your_table_name` with the actual table name)
df = pd.read_sql_query("""SELECT * FROM Fires""", conn)

# Close the connection
#conn.close()

# Display the first few rows
df.head()

```

```python
gregorian_dates = pd.to_datetime(df['DISCOVERY_DATE'], origin='julian', unit='D')

# Filter for dates from 2009 onward
df_filtered = df[gregorian_dates.dt.year >= 2009].copy()
df_filtered['Year'] = gregorian_dates.dt.year

df_filtered['Month'] = gregorian_dates.dt.month

# Count occurrences per year and month, then average per year
avg_monthly_fires = df_filtered.groupby('Year')['Month'].count() / 12

# Convert to integer for readability
avg_monthly_fires = avg_monthly_fires.round().astype(int)

print(avg_monthly_fires)
```

```python
"Average Per All Months"

# Convert to datetime
gregorian_dates = pd.to_datetime(df['DISCOVERY_DATE'], origin='julian', unit='D')

# Filter dates from 2009 onward
filtered_dates = gregorian_dates[gregorian_dates.dt.year >= 2009]

# Get value counts for months
filtered_dates.dt.month.value_counts().sort_index()

total_sum = filtered_dates.dt.month.value_counts().sum()

average_per_month = total_sum/72

average_per_month

```
