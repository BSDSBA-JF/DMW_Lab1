```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)

# Connect to the SQLite database file
conn = sqlite3.connect("FPA_FOD_20170508.sqlite")

# Read a table into a Pandas DataFrame (replace `your_table_name` with the actual table name)
df = pd.read_sql_query("""SELECT * FROM Fires""", conn)

# Close the connection
#conn.close()

# Display the first few rows
df.head()

```

```python
filt = df['STATE'] == 'OR'

df_filt = df[filt]

df_filt
```

```python
# Extract and count occurrences of first two digits
df['Hours'] = df_filt['DISCOVERY_TIME'].astype(str).str[:2]
count_series = df['Hours'].value_counts().sort_index()

# Create a full range of "00" to "99"
full_range = pd.Series(0, index=[f"{i:02}" for i in range(24)])  # Creates "00" to "99" with 0 count

# Merge the full range with actual counts, filling missing ones with 0
count_series = full_range.add(count_series, fill_value=0).astype(int)

# Convert to DataFrame for plotting
count_df = count_series.reset_index()
count_df.columns = ['Hours', 'Fire Count']

filt = count_df['Hours'] != 'No'
df = count_df[filt]

df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Hours'], df['Fire Count'], marker='o', linestyle='-', color='red')

# Labels and title
plt.xlabel("Hours")
plt.ylabel("Fire Count")
plt.title("Fires Occurrences by the Hour")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(True)

# Ensure y-axis starts at 0
plt.ylim(0, max(df['Fire Count']) + 500)  

# Show the plot
plt.show()
```
