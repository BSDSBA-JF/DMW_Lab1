<h1 style="color:#ffffff; background-color:#222f3e; border: 1px solid #ffffff; padding: 10px 0;">II. PROBLEM STATEMENT</h1>

Wildfires are a growing crisis in the United States, causing widespread destruction and straining firefighting resources. At the same time, house fires remain a threat, requiring immediate response to prevent loss of life and property. Fire departments must balance their resources to effectively combat both types of fires, ensuring that wildfire response does not compromise urban fire safety.
    
This study aims to analyze historical fire data from Oregon wildfires and London house fires to identify patterns in time of day and seasonal trends, which will help Oregon fire stations optimize preparedness.

<a id='Motivation'></a>
<h1 style="color:#ffffff; background-color:#222f3e; border: 1px solid #ffffff; padding: 10px 0;">III. MOTIVATION</h1>

The reason for incorporating London house fire data is due to the similar weather patterns and climate between Oregon and London, particularly their mild, wet winters and dry summers. By understanding how weather and seasonal trends impact house fire occurrences in London, we can apply similar insights to predict high-risk periods for house fires in Oregon.
                                                                                                                                
Specifically, this study can provide insights into:
1. Water conservation strategies to ensure that there are enough reserves left for wildfire-prone seasons,
2. Budget planning, such as funding for emergency response efforts and firefighting equipment,
3. Resource Rental to ensure support, such as aerial support and helicopters, is available during peak wildfire season,
4. Staffing adjustments to allow fire stations to increase personnel during high-risk periods.
                                                                            
With this, Oregon can enhance its fire response strategies, improve resource allocation, and reduce fire damage and loss of life.
